```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=5, kernel_size=4, stride=2)
        self.conv2 = nn.Conv2d(in_channels=5, out_channels=5, kernel_size=3, stride=2)
        self.conv3 = nn.Conv2d(in_channels=5, out_channels=5, kernel_size=2, stride=1)
        self.dropout = nn.Dropout()
        self.fc1 = nn.Linear(5*30*30, 30*30)
        self.fcfinal = nn.Linear(30*30, 1)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = self.dropout(x)
        x = F.relu(self.conv3(x))
        x = x.view(-1, 5*30*30)
        x = F.relu(self.fc1(x))
        out = self.fcfinal(x)
        return out

# training
model = Net()
num_epochs = 100
learning_rate = 1e-4
lambda_reg = 1e-3
iters, losses = [], []
iters_sub, train_acc, val_acc = [], [] ,[]

optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.MSELoss()
itercount = 0 # the number of iterations

for epoch in range(num_epochs):
    for sample, label in iter(trainloader):
        prediction = model(sample)
        weights = torch.cat([x.view(-1) for x in model.parameters()])
        loss = criterion(prediction, label)
        
        # add regularization
        l2_reg = None
        for W in model.parameters():
            if l2_reg is None:
                l2_reg = W.norm(2)
            else:
                l2_reg = l2_reg + W.norm(2)
        loss = loss + lambda_reg * l2_reg
        
        loss.backward() # backprop
        optimizer.step() 
        optimizer.zero_grad()
        iters.append(itercount)
        losses.append(float(loss)/batch_size) # compute *average* loss
        if itercount % 10 == 0:
            iters_sub.append(itercount)
            train_acc.append(get_accuracy(model, trainset))
            val_acc.append(get_accuracy(model, valset))
        if itercount % 100 == 0:
            print("-"*10, "Iteration:", itercount, "-"*10)
            print("Loss:", float(loss)/batch_size)
            print("Training accuracy:", get_accuracy(model, trainset))
            print("Validation accuracy:", get_accuracy(model, valset))
            print("="*50)
        # increment the iteration number
        itercount += 1
```

```python
# plotting
plt.title("Training Curve".format(batch_size, learning_rate))
plt.plot(iters, losses, label="Train")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.show()
plt.title("Training Curve".format(batch_size, learning_rate))
plt.plot(iters_sub, train_acc, label="Train")
plt.plot(iters_sub, val_acc, label="Validation")
plt.xlabel("Iterations")
plt.ylabel("Accuracy")
plt.legend(loc='best')
plt.show()

print("Training accuracy is: ", get_accuracy(model, trainset))
print("Validation accuracy is: ", get_accuracy(model, valset))
```
