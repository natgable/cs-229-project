This notebook serves to preprocess all our data by turning the jpg images to 3-channel numpy arrays.

```python
import os
import glob

import tqdm
import numpy as np
import scipy.ndimage
import scipy.misc
import sklearn.metrics

from PIL import Image

import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd

from collections import defaultdict

import random

import argparse
IMG_SIZE = (128,128)
```

```python
# load data as numpy array
for image_path in tqdm.tqdm(list(glob.glob('./Post-hurricane/***/**/*.jpeg'))):
    image = Image.open(image_path)
    image_pixels = np.array(image)
    image_basepath, _ = os.path.splitext(image_path)
    np.savez(image_basepath+'.npz', pixels=image_pixels)
```

```python
# test out a sample of the images and the numpy array
sample_damage_path = "./Post-hurricane/train_another/damage/-93.743629_30.073390999999997"
sample_no_damage_path = "./Post-hurricane/train_another/no_damage/-95.215981_30.0308"

sample_im_damage = Image.open(sample_damage_path + ".jpeg")
plt.imshow(sample_im_damage)
plt.title("Image with damage")
plt.show()

sample_np_damage = np.load(sample_damage_path + ".npz")
print("Image shape:", sample_np_damage["pixels"].shape)

sample_im_no_damage = Image.open(sample_no_damage_path + ".jpeg")
plt.imshow(sample_im_no_damage)
plt.title("Image with no damage")
plt.show()

sample_np_no_damage = np.load(sample_no_damage_path + ".npz")
print("Image shape:", sample_np_no_damage["pixels"].shape)
```
